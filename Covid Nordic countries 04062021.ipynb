```python
# COVID-19 SITUATION IN FINLAND, SWEDEN AND NORWAY 4.6.2021

# The objective was to get a quick update on Covid situation in the Finland, Sweden and Norway

# I used a dataset 1.1.2020 - 2.6.2021: JHU CSSE COVID-19 Dataset 2021-06-03

# The Dataset was downloaded on June, 4th, 2021.

# Analysis was done right away June, 4th, 2021.

# Data can be found https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data

from datetime import datetime
print(f'Päivitetty {datetime.now()}')
```

```python
# Libraries

import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))

import matplotlib.dates as mdates

plt.style.use('seaborn-whitegrid')
```

```python
# Datasets

finland = pd.read_excel('Finland.xlsx')
sweden = pd.read_excel('Sweden.xlsx')
norway = pd.read_excel('Norway.xlsx')
```

```python
# Index

finland.index=pd.to_datetime(finland['Date'],format='%Y-%m-%d')
finland=finland.drop('Date',axis=1)
```

```python
# Index

sweden.index=pd.to_datetime(sweden['Date'],format='%Y-%m-%d')
sweden=sweden.drop('Date',axis=1)
```

```python
# Index

norway.index=pd.to_datetime(norway['Date'],format='%Y-%m-%d')
norway=norway.drop('Date',axis=1)
```

```python
# Plotting all three countries, checking death rate

import matplotlib.pyplot as plt
import pandas as pd

finland = pd.read_excel('Finland.xlsx')
sweden = pd.read_excel('Sweden.xlsx')
norway = pd.read_excel('Norway.xlsx')

# gca stands for 'get current axis'
ax = plt.gca()

finland.plot(kind='line',x='Date',y='Deaths in Finland', color='blue', ax=ax)
sweden.plot(kind='line',x='Date',y='Deaths in Sweden', color='yellow', ax=ax)
norway.plot(kind='line',x='Date',y='Deaths in Norway', color='red', ax=ax)

plt.show()
```
